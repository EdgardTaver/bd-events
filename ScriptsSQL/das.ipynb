```python
import random

def random_number_of_size(n):
    sequence = ""
    
    while len(sequence) < n:
        random_num = random.randint(0,9)
        sequence += str(random_num)
    
    return sequence

random_number_of_size(10)
```

```python
n = 0

while n < 15:
    s = random_number_of_size(33)
    print(s)
    n += 1
```

```python
def random_phone():
    sequence = "("
    
    sequence += str(random.randint(0,9))
    sequence += str(random.randint(0,9))
    
    sequence += ")9"
    
    n = 0
    while n < 4:
        sequence += str(random.randint(0,9))
        n += 1
        
    sequence += "-"
    
    n = 0
    while n < 4:
        sequence += str(random.randint(0,9))
        n += 1
    
    return sequence

n = 0
while n < 30:
    s = random_phone()
    print(s)
    n += 1   
```
